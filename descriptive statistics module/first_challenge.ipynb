### What is the mean, median, 25th percentile, 75th percentile, minimum, and maximum for each of the numeric columns?

```python
columns = [
    'Income', 'Kidhome', 'Teenhome', 'Recency', 'MntWines',
    'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts',
    'MntGoldProds', 'NumDealsPurchases', 'NumWebPurchases',
    'NumCatalogPurchases', 'NumStorePurchases', 'NumWebVisitsMonth',
    'AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5', 'AcceptedCmp1',
    'AcceptedCmp2', 'Complain', 'Z_CostContact', 'Z_Revenue', 'Response',
    'Age', 'Customer_Days', 'marital_Divorced', 'marital_Married',
    'marital_Single', 'marital_Together', 'marital_Widow',
    'education_2n Cycle', 'education_Basic', 'education_Graduation',
    'education_Master', 'education_PhD', 'MntTotal', 'MntRegularProds',
    'AcceptedCmpOverall', 'kids', 'expenses'
]

for column in columns:
    mean_value = data[column].mean()
    median_value = data[column].median()
    mode_value = data[column].mode().values[0]  # Get the first mode if there are multiple
    max_value = np.max(data[column])
    min_value = np.min(data[column])
    percentile_25 = np.percentile(data[column], 25)
    percentile_75 = np.percentile(data[column], 75)

    print(f"\n{column}:")
    print(f"mean: {mean_value:.2f}, median: {median_value:.2f}, mode: {mode_value:.2f}")
    print(f"max: {max_value:.2f}, min: {min_value:.2f}")
    print(f"25th: {percentile_25:.2f}, 75th: {percentile_75:.2f}")
```

### Do our clients have higher or lower levels of education? Look in the education_level column


```python
# Most clients have a undergradute degree. A non-negligible number of clients are Masters and PhDs. A very well-educated cohort.
# Disclaimer: this is the third iteration of the code. In the first versions, the fact that multiple clients had multiple
# levels of education recorded was messing with the calculations. Here, I created a new education column that takes the
# highest level of education for each customer and calculated the distribution based on this new column.

# List of education columns
education_columns = ['education_2n Cycle', 'education_Basic', 'education_Graduation', 'education_Master', 'education_PhD']

# Check for multiple education levels per customer
education_sum = data[education_columns].sum(axis=1)

print(f"Total customers: {len(data)}")
print(f"Customers with no education level marked: {(education_sum == 0).sum()}")
print(f"Customers with exactly one education level: {(education_sum == 1).sum()}")
print(f"Customers with multiple education levels: {(education_sum > 1).sum()}")

# Create a new column with the education level
data['education'] = data[education_columns].idxmax(axis=1).str.replace('education_', '')

# Replace empty string with 'Unknown' for customers with no education level marked
data.loc[education_sum == 0, 'education'] = 'Unknown'

# Count the occurrences of each education level
education_counts = data['education'].value_counts()

# Calculate percentages
education_percentages = (education_counts / len(data) * 100).round(2)

# Create a DataFrame with counts and percentages
education_summary = pd.DataFrame({
    'Count': education_counts,
    'Percentage': education_percentages
})

# Display the summary
print("\nEducation Level Summary:")
print(education_summary)

# Visualize the distribution
plt.figure(figsize=(10, 6))
education_summary['Percentage'].plot(kind='bar')
plt.title('Distribution of Customer Education Levels')
plt.xlabel('Education Level')
plt.ylabel('Percentage of Customers')
plt.xticks(rotation=45, ha='right')
for i, v in enumerate(education_summary['Percentage']):
    plt.text(i, v, f'{v}%', ha='center', va='bottom')
plt.tight_layout()
plt.show()
```
