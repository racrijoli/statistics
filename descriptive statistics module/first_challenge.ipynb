### What is the mean, median, 25th percentile, 75th percentile, minimum, and maximum for each of the numeric columns?

```python
columns = [
    'Income', 'Kidhome', 'Teenhome', 'Recency', 'MntWines',
    'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts',
    'MntGoldProds', 'NumDealsPurchases', 'NumWebPurchases',
    'NumCatalogPurchases', 'NumStorePurchases', 'NumWebVisitsMonth',
    'AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5', 'AcceptedCmp1',
    'AcceptedCmp2', 'Complain', 'Z_CostContact', 'Z_Revenue', 'Response',
    'Age', 'Customer_Days', 'marital_Divorced', 'marital_Married',
    'marital_Single', 'marital_Together', 'marital_Widow',
    'education_2n Cycle', 'education_Basic', 'education_Graduation',
    'education_Master', 'education_PhD', 'MntTotal', 'MntRegularProds',
    'AcceptedCmpOverall', 'kids', 'expenses'
]

for column in columns:
    mean_value = data[column].mean()
    median_value = data[column].median()
    mode_value = data[column].mode().values[0]  # Get the first mode if there are multiple
    max_value = np.max(data[column])
    min_value = np.min(data[column])
    percentile_25 = np.percentile(data[column], 25)
    percentile_75 = np.percentile(data[column], 75)

    print(f"\n{column}:")
    print(f"mean: {mean_value:.2f}, median: {median_value:.2f}, mode: {mode_value:.2f}")
    print(f"max: {max_value:.2f}, min: {min_value:.2f}")
    print(f"25th: {percentile_25:.2f}, 75th: {percentile_75:.2f}")
```

### Do our clients have higher or lower levels of education? Look in the education_level column


```python
# Most clients have a undergradute degree. A non-negligible number of clients are Masters and PhDs. A very well-educated cohort.
# Disclaimer: this is the third iteration of the code. In the first versions, the fact that multiple clients had multiple
# levels of education recorded was messing with the calculations. Here, I created a new education column that takes the
# highest level of education for each customer and calculated the distribution based on this new column.

# List of education columns
education_columns = ['education_2n Cycle', 'education_Basic', 'education_Graduation', 'education_Master', 'education_PhD']

# Check for multiple education levels per customer
education_sum = data[education_columns].sum(axis=1)

print(f"Total customers: {len(data)}")
print(f"Customers with no education level marked: {(education_sum == 0).sum()}")
print(f"Customers with exactly one education level: {(education_sum == 1).sum()}")
print(f"Customers with multiple education levels: {(education_sum > 1).sum()}")

# Create a new column with the education level
data['education'] = data[education_columns].idxmax(axis=1).str.replace('education_', '')

# Replace empty string with 'Unknown' for customers with no education level marked
data.loc[education_sum == 0, 'education'] = 'Unknown'

# Count the occurrences of each education level
education_counts = data['education'].value_counts()

# Calculate percentages
education_percentages = (education_counts / len(data) * 100).round(2)

# Create a DataFrame with counts and percentages
education_summary = pd.DataFrame({
    'Count': education_counts,
    'Percentage': education_percentages
})

# Display the summary
print("\nEducation Level Summary:")
print(education_summary)

# Visualize the distribution
plt.figure(figsize=(10, 6))
education_summary['Percentage'].plot(kind='bar')
plt.title('Distribution of Customer Education Levels')
plt.xlabel('Education Level')
plt.ylabel('Percentage of Customers')
plt.xticks(rotation=45, ha='right')
for i, v in enumerate(education_summary['Percentage']):
    plt.text(i, v, f'{v}%', ha='center', va='bottom')
plt.tight_layout()
plt.show()
```

### How many clients do we have in each marital status? Look in the marital_status column

```python
# Here we go.

# List of marital status columns
marital_columns = ['marital_Divorced', 'marital_Married', 'marital_Single', 'marital_Together', 'marital_Widow']

# Check for multiple marital statuses per customer
marital_sum = data[marital_columns].sum(axis=1)

print(f"Total customers: {len(data)}")
print(f"Customers with no marital status marked: {(marital_sum == 0).sum()}")
print(f"Customers with exactly one marital status: {(marital_sum == 1).sum()}")
print(f"Customers with multiple marital statuses: {(marital_sum > 1).sum()}")

# Create a new column with the marital status
data['marital_status'] = data[marital_columns].idxmax(axis=1).str.replace('marital_', '')

# Replace empty string with 'Unknown' for customers with no marital status marked
data.loc[marital_sum == 0, 'marital_status'] = 'Unknown'

# Count the occurrences of each marital status
marital_counts = data['marital_status'].value_counts()

# Calculate percentages
marital_percentages = (marital_counts / len(data) * 100).round(2)

# Create a DataFrame with counts and percentages
marital_summary = pd.DataFrame({
    'Count': marital_counts,
    'Percentage': marital_percentages
})

# Display the summary
print("\nMarital Status Summary:")
print(marital_summary)

# Visualize the distribution
plt.figure(figsize=(10, 6))
marital_summary['Percentage'].plot(kind='bar')
plt.title('Distribution of Customer Marital Statuses')
plt.xlabel('Marital Status')
plt.ylabel('Percentage of Customers')
plt.xticks(rotation=45, ha='right')
for i, v in enumerate(marital_summary['Percentage']):
    plt.text(i, v, f'{v}%', ha='center', va='bottom')
plt.tight_layout()
plt.show()
```

### What is the relationship between marital status and number of children? Do married people have more children? Children is in the kids column


```python
# Let's see. Seems like divorced people and people in relationships have more kids. But to be fair, the difference is not that big.

# First, ensure we have a single marital status for each customer
marital_columns = ['marital_Divorced', 'marital_Married', 'marital_Single', 'marital_Together', 'marital_Widow']
data['marital_status'] = data[marital_columns].idxmax(axis=1).str.replace('marital_', '')

# Calculate total number of children (assuming 'Kidhome' is for young children and 'Teenhome' for teenagers)
data['total_children'] = data['Kidhome'] + data['Teenhome']

# Calculate average number of children for each marital status
avg_children = data.groupby('marital_status')['total_children'].mean().sort_values(ascending=False)

print("Average number of children by marital status:")
print(avg_children)

# Visualize the relationship
plt.figure(figsize=(10, 6))
sns.boxplot(x='marital_status', y='total_children', data=data)
plt.title('Distribution of Number of Children by Marital Status')
plt.xlabel('Marital Status')
plt.ylabel('Number of Children')
plt.show()

# Calculate percentage of people with children for each marital status
pct_with_children = (data.groupby('marital_status')['total_children'].apply(lambda x: (x > 0).mean()) * 100).sort_values(ascending=False)

print("\nPercentage of people with children by marital status:")
print(pct_with_children)
```

### Do people spend more or less on our platform when they have children? See the expenses and kids columns

```python
# This is interesting: people with no children spend more on the platform! Significantly more.

# Calculate total number of children
data['total_children'] = data['Kidhome'] + data['Teenhome']

# Create a binary variable for having children or not
data['has_children'] = data['total_children'] > 0

# Calculate total spending (adjust these columns based on your actual spending columns)
spending_columns = ['MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']
data['total_spending'] = data[spending_columns].sum(axis=1)

# Calculate average spending for people with and without children
avg_spending = data.groupby('has_children')['total_spending'].mean()

print("Average spending:")
print(avg_spending)

# Visualize the distribution of spending
plt.figure(figsize=(10, 6))
sns.boxplot(x='has_children', y='total_spending', data=data)
plt.title('Distribution of Total Spending by Presence of Children')
plt.xlabel('Has Children')
plt.ylabel('Total Spending')
plt.show()
```


### Do people who have a higher salary spend more? See the Income and expenses columns

```python
# As expected, people who earn more spend more.

# Calculate total spending (adjust these columns based on your actual spending columns)
spending_columns = ['MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']
data['total_spending'] = data[spending_columns].sum(axis=1)

# Visualize the relationship between income and spending
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Income', y='total_spending', data=data)
plt.title('Relationship between Income and Total Spending')
plt.xlabel('Income')
plt.ylabel('Total Spending')
plt.show()

# Group data into income brackets and calculate average spending
data['income_bracket'] = pd.cut(data['Income'], bins=5, labels=['Very Low', 'Low', 'Medium', 'High', 'Very High'])
avg_spending_by_bracket = data.groupby('income_bracket')['total_spending'].mean().sort_values(ascending=True)

print("\nAverage spending by income bracket:")
print(avg_spending_by_bracket)

# Visualize average spending by income bracket
plt.figure(figsize=(10, 6))
avg_spending_by_bracket.plot(kind='bar')
plt.title('Average Spending by Income Bracket')
plt.xlabel('Income Bracket')
plt.ylabel('Average Total Spending')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Married people spend the least in our platform. Widowers are the ones who spend the most.

# First, ensure we have a single marital status for each customer
marital_columns = ['marital_Divorced', 'marital_Married', 'marital_Single', 'marital_Together', 'marital_Widow']
data['marital_status'] = data[marital_columns].idxmax(axis=1).str.replace('marital_', '')

# Calculate total spending (adjust these columns based on your actual spending columns)
spending_columns = ['MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']
data['total_spending'] = data[spending_columns].sum(axis=1)

# Create a binary variable for relationship status
data['in_relationship'] = data['marital_status'].isin(['Married', 'Together'])

# Calculate average spending for people in relationships and single
avg_spending = data.groupby('in_relationship')['total_spending'].mean()

print("Average spending:")
print(avg_spending)

# Visualize the distribution of spending
plt.figure(figsize=(10, 6))
sns.boxplot(x='in_relationship', y='total_spending', data=data)
plt.title('Distribution of Total Spending by Relationship Status')
plt.xlabel('In Relationship')
plt.ylabel('Total Spending')
plt.xticks([0, 1], ['Single', 'In Relationship'])
plt.show()

# Calculate average spending for each marital status
avg_spending_by_status = data.groupby('marital_status')['total_spending'].mean().sort_values(ascending=False)

print("\nAverage spending by marital status:")
print(avg_spending_by_status)

# Visualize average spending by marital status
plt.figure(figsize=(10, 6))
avg_spending_by_status.plot(kind='bar')
plt.title('Average Spending by Marital Status')
plt.xlabel('Marital Status')
plt.ylabel('Average Total Spending')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
